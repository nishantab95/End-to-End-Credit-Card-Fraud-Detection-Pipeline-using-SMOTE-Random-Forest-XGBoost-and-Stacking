```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
try:
  df = pd.read_csv('/content/drive/MyDrive/Data sets/credit card ml project/cleaned_data/clean_data.csv')    # Google Drive (Colab)
except FileNotFoundError:
  df = pd.read_csv("E:/AI ML/Project/Data/clean_data.csv")   # local drive
```

```python
# imports
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
```

```python
# defining x and y
x = df.drop('Class', axis=1)
y = df['Class']
```

```python
# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42, stratify=y)
```

```python
# scaling the data
scaler = RobustScaler()
x_train_robust = scaler.fit_transform(x_train)
x_test_robust = scaler.transform(x_test)
```

```python
# using smote to balance the data
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_train_sm, y_train_sm = smote.fit_resample(x_train_robust, y_train)
```

```python
# saving x and y data

import os
import joblib

data_dict = {
    "X_train": x_train_sm,
    "y_train": y_train_sm,
    "X_test": x_test_robust,
    "y_test": y_test
}

# try saving to Google Drive (Colab)
try:
    save_dir = "/content/drive/MyDrive/AI ML/Project/Preprocessed/artifacts"
    os.makedirs(save_dir, exist_ok=True)

    save_path = os.path.join(save_dir, "preprocessed_data.pkl")

    joblib.dump(data_dict, save_path)

    print("Saved to Google Drive:", save_path)

except Exception as e:

    print("Not running in Colab, saving locally...")

    # local VSCode save
    save_dir = "E:/AI ML/Project/Data/artifacts"
    os.makedirs(save_dir, exist_ok=True)

    save_path = os.path.join(save_dir, "preprocessed_data.pkl")

    joblib.dump(data_dict, save_path)

    print("Saved locally:", save_path)
```
